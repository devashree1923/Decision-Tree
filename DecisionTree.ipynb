***Import packages and classes***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO

from IPython.display import Image
import pydotplus
```

```python
dataset = pd.read_csv("heart.csv")
dataset.head()
```

```python
x = dataset.iloc[:, :-1]
y = dataset.iloc[:,13]
x = np.array(x)
y = np.array(y)
```

```python
x
```

```python
dataset.shape
```

```python
print(x.shape)
print(y.shape)
```

***Decision Tree Classifier***

```python
tree = DecisionTreeClassifier(criterion='entropy')
# tree = DecisionTreeClassifier(criterion='entropy', max_depth=4)
# tree = DecisionTreeClassifier(criterion='entropy', min_samples_leaf=50)
# tree = DecisionTreeClassifier(criterion='entropy', min_samples_split=10)
tree.fit(x,y)
```

***Plotting Decisison Tree***

```python
dot_data = StringIO()
export_graphviz(tree, out_file=dot_data,
                filled=True, rounded = True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png(), width=1500, height=800)
```

***Predicting***

```python
predict = tree.predict(x)
predict
```

***Metrics***

```python
from sklearn import metrics
print("Accuracy: ", metrics.accuracy_score(y,predict))
```
